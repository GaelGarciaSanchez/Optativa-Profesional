Importar librerias

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
```

Graficos de lineas


```python
plt.figure(figsize=(10,4))
plt.plot(x,y)
plt.title("Grafico de linea")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

Generar vectores  para un grafico de barras

```python
categorias = ['A','B','C','D']
valores = [3,7,2,5]

plt.figure(figsize=(10,4))
plt.bar(categorias,valores)
plt.title("Grafico de linea")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

Grafico de  Dispercion

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.figure(figsize=(10,4))
plt.scatter(x,y)
plt.title("Grafico de Dispercion")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

Histogramas

```python
datos = np.random.randn(1000)

plt.figure(figsize=(10,4))
plt.hist(datos,bins=30)
plt.title("Histograma")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.show()
```
